# Student Progress Tracker with Insights and Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Collect User Input
def get_student_data():
    print("Welcome to the Enhanced Student Progress Tracker!")
    student_name = input("Enter the student's name: ")
    course_name = input("Enter the course name: ")
    total_modules = int(input("Enter the total number of modules in the course: "))
    completed_modules = int(input("Enter the number of modules completed: "))
    
    time_spent = []
    print("Enter time spent (in hours) for each completed module:")
    for i in range(completed_modules):
        time = float(input(f"Module {i+1}: "))
        time_spent.append(time)
    
    return student_name, course_name, total_modules, completed_modules, time_spent

# Step 2: Calculate Insights and Determine Pass Status
def calculate_insights(total_modules, completed_modules, time_spent, avg_time_per_module=5):
    progress_percentage = (completed_modules / total_modules) * 100
    avg_time_actual = sum(time_spent) / completed_modules if completed_modules > 0 else 0
    
    strengths = []
    weaknesses = []
    passed_modules = 0
    
    for i, time in enumerate(time_spent):
        if time >= avg_time_per_module:
            strengths.append((f"Module {i+1}", time))
            passed_modules += 1
        else:
            weaknesses.append((f"Module {i+1}", time))
    
    # Pass Criteria
    passed_completion = progress_percentage >= 50
    passed_time = passed_modules >= (completed_modules / 2)
    passed = passed_completion and passed_time

    return progress_percentage, avg_time_actual, strengths, weaknesses, passed

# Step 3: Display Insights and Pass/Fail Outcome
def display_insights(student_name, course_name, progress_percentage, avg_time_actual, strengths, weaknesses, passed):
    print("\n=== Progress and Insights Report ===")
    print(f"Student: {student_name}")
    print(f"Course: {course_name}")
    print(f"Progress: {progress_percentage:.2f}% completed")
    print(f"Average Time Spent per Module: {avg_time_actual:.2f} hours")
    
    print("\nStrengths (Modules where the student excelled):")
    for module, time in strengths:
        print(f"  - {module}: {time:.2f} hours")
    
    print("\nWeaknesses (Modules where the student struggled):")
    for module, time in weaknesses:
        print(f"  - {module}: {time:.2f} hours")
    
    # Pass or Fail Message
    if passed:
        print("\nCongratulations! You have passed the course requirements. ðŸŽ‰")
    else:
        print("\nYou did not meet the course requirements. Keep trying and you'll succeed next time!")

# Step 4: Visualize Insights with a Beautified Chart
def visualize_insights(time_spent, avg_time_per_module):
    modules = [f"Module {i+1}" for i in range(len(time_spent))]
    expected_time = [avg_time_per_module] * len(time_spent)
    
    fig, ax = plt.subplots(figsize=(10, 5))

    # Bar Chart Comparing Actual vs Expected Time
    bars = ax.bar(modules, time_spent, label='Actual Time Spent', color=plt.cm.viridis(np.linspace(0, 1, len(time_spent))), alpha=0.8)

    # Plotting the expected time line
    ax.plot(modules, expected_time, label='Expected Time (Model)', color='#FF6347', linestyle='--', linewidth=3)

    # Adding text annotations to the bars
    for i, bar in enumerate(bars):
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.2, f'{yval:.2f}h', ha='center', va='bottom', fontsize=10, color='black')

    # Customize labels and title
    ax.set_xlabel("Modules", fontsize=14, fontweight='bold', family='Arial')
    ax.set_ylabel("Time Spent (hours)", fontsize=14, fontweight='bold', family='Arial')
    ax.set_title("Student Performance: Actual vs Expected Time", fontsize=16, fontweight='bold', family='Arial')
    
    # Adding gridlines for better readability
    ax.grid(True, axis='y', linestyle='--', alpha=0.5)
    
    # Adding a legend
    ax.legend(fontsize=12)
    
    # Adjust layout to avoid overlap
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Main function to run the program
def main():
    student_name, course_name, total_modules, completed_modules, time_spent = get_student_data()
    progress_percentage, avg_time_actual, strengths, weaknesses, passed = calculate_insights(
        total_modules, completed_modules, time_spent
    )
    display_insights(student_name, course_name, progress_percentage, avg_time_actual, strengths, weaknesses, passed)
    visualize_insights(time_spent, avg_time_per_module=5)

# Run the main function
if __name__ == "__main__":
    main()
```
